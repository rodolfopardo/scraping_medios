```python
#Se limpian los datos 

def limpieza_datos(x):
    x['tweet'] = x['tweet'].str.replace(r'https?:\/\/.*[\r\n]*', '')
    x['tweet'] = x['tweet'].str.replace(r'pic.twitter.com/.*', '')
    return x

clarin = limpieza_datos(clarin)
elcomercio_pe = limpieza_datos(elcomercio_pe)
elmercurio = limpieza_datos(elmercurio)
elmundoes = limpieza_datos(elmundoes)
elpaisespana = limpieza_datos(elpaisespana)
eltelegrafo = limpieza_datos(eltelegrafo)
eltiempo = limpieza_datos(eltiempo)
eluniversalctg = limpieza_datos(eluniversalctg)
eluniversalmx = limpieza_datos(eluniversalmx)
eluniversocom = limpieza_datos(eluniversocom)
folha = limpieza_datos(folha)
ilmessa = limpieza_datos(ilmessa)
jornada = limpieza_datos(jornada)
lanacion = limpieza_datos(lanacion)
larepublicaperu = limpieza_datos(larepublicaperu)
lasegundaperu = limpieza_datos(lasegundaperu)
nytimes = limpieza_datos(nytimes)
oglobo = limpieza_datos(oglobo)
repubblica = limpieza_datos(repubblica)
wp = limpieza_datos(wp)
```

```python
texto = wp['tweet'].to_list()
stop = stop_words.get_stop_words("es")
with open('texto.txt', 'w') as f:
    for item in texto:
        f.write("%s\n" % item)
stylecloud.gen_stylecloud(file_path='texto.txt',
                          icon_name='fab fa-twitter',
                          custom_stopwords=stop, 
                          output_name='wp.png')
```
